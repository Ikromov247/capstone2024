```python
from ultralytics import YOLO

# Load a pretrained YOLOv8m model
model = YOLO('yolov8m.pt')

# Open an image using PIL
source = "/content/drive/MyDrive/capstone_data/capstone2024/human_detection_dataset/human"

# Run inference on the source
results = model(source)  # list of Results objects
```

```python
results[0]
```

```python
import matplotlib.pyplot as plt
import random

num_images = len(results)
fig, ax = plt.subplots(8, 4, figsize=(12,16))

indeces = random.choices( [x for x in range(0, num_images)], k=32)

indx, indy = 0, 0
for i in indeces:
  if indx==8:
      indy+=1
      indx=0
  ax[indx, indy].imshow(results[i].plot())
  indx+=1


plt.show()
```
